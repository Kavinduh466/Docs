```python
import pandas as pd

def null_counts(df):
    """
    Prints the number of null values for each column in the DataFrame
    """
    for col in df.columns:
        null_count = df[col].isnull().sum()
        print(f"{col}: {null_count}")
        

null_counts(df)
```

```python
# Clean df7 into df_cleaned

df_cleaned = df7.copy()

# 1. Remove opposite sign rows
df_cleaned = df_cleaned[~(
    ((df_cleaned["Inflation_Rate_Percent"] > 0) & (df_cleaned["Currency_Change_YTD_Percent"] < 0)) |
    ((df_cleaned["Inflation_Rate_Percent"] < 0) & (df_cleaned["Currency_Change_YTD_Percent"] > 0))
)]

# 2. Calculate averages
infl_avg = df_cleaned["Inflation_Rate_Percent"].mean()
curr_avg = df_cleaned["Currency_Change_YTD_Percent"].mean()

# 3. Remove rows where both values are above their averages
df_cleaned = df_cleaned[~(
    (df_cleaned["Inflation_Rate_Percent"] > infl_avg) &
    (df_cleaned["Currency_Change_YTD_Percent"] > curr_avg)
)]

print("Rows before cleaning:", len(df7))
print("Rows after cleaning:", len(df_cleaned))
print("Rows removed:", len(df7) - len(df_cleaned))
```

```python
x = df_cleaned.drop('Inflation_Rate_Percent',axis=1)
y = df_cleaned.Inflation_Rate_Percent
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(x_train,y_train)
lr_clf.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5,test_size=0.1, random_state=0)
cross_val_score(LinearRegression(), x,y,cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['squared_error','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(x,y)
```
