```python
import matplotlib.pyplot as plt

# Clean column names
df.columns = df.columns.str.strip()

categorical_cols = [
    "Do you have any bad habits like smoking, drinking on a daily basis?",
    "Your Academic Stage",
    "Study Environment",
    "What coping strategy you use as a student?",
    "Peer pressure"
]

target_col = "Rate your academic stress index"

# Typical markers for categorical plotting
markers = ['+', 'x', '*', '1', '2', '3', '4', '|', '_']

for col in categorical_cols:
    plt.figure(figsize=(8,5))
    
    categories = df[col].dropna().unique()
    
    for i, category in enumerate(categories):
        subset = df[df[col] == category]
        plt.scatter(
            subset.index,
            subset[target_col],
            label=category,
            marker=markers[i % len(markers)],  # cycle markers
            s=100,                             # marker size
            edgecolors='black'                 # outline for better visibility
        )
    
    plt.title(f"{col} vs {target_col}")
    plt.xlabel("Student Index")
    plt.ylabel("Academic Stress Index")
    plt.legend(title=col, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

```

```python
x = df_encoded.drop('Rate your academic stress index ', axis='columns')
y = df_encoded['Rate your academic stress index ']; 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(x_train,y_train)
lr_clf.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5,test_size=0.2, random_state=0)
cross_val_score(LinearRegression(), x,y,cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['squared_error','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(x,y)
```
