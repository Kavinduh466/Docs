```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import json

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

try:
    url = "http://quotes.toscrape.com/"
    driver.get(url)

    quote_elements = driver.find_elements(By.CLASS_NAME, "quote")

    all_quotes = []

    for idx, quote in enumerate(quote_elements, start=1):
        # Extract text, author, and visual/rendered properties
        text_element = quote.find_element(By.CLASS_NAME, "text")
        author_element = quote.find_element(By.CLASS_NAME, "author")

        quote_data = {
            "quote_index": idx,
            "text": text_element.text,
            "author": author_element.text,
            "font_size": text_element.value_of_css_property("font-size"),
            "color": text_element.value_of_css_property("color"),
            "location": text_element.location,  # x, y coordinates
            "size": text_element.size           # height, width
        }
        all_quotes.append(quote_data)

    # --- Print all extracted quotes ---
    print(json.dumps(all_quotes, indent=4))

finally:
    # --- Close the browser ---
    driver.quit()
```
